```python
import matplotlib.pyplot as plt
import numpy as np

lst=[91, 29, 64, 7]
sm=sum(lst)

a=round((91 / sm) * 100, 1)  # üëâÔ∏è 50.8
b=round((29 / sm) * 100, 1)  # üëâÔ∏è 50.8
c=round((64 / sm) * 100, 1)  # üëâÔ∏è 50.8
d=round((7 / sm) * 100, 1)  # üëâÔ∏è 50.8

y = np.array([a, b, c, d])
labels = ["Majoritarian", "Combined", "PR", "No Directed Elections"]

plt.pie(y, labels = labels)
plt.legend()
plt.show() 
```

```python
import matplotlib.pyplot as plt
plt.rc('font', size=15)        # Í∏∞Î≥∏ Ìè∞Ìä∏ ÌÅ¨Í∏∞  https://wooono.tistory.com/271

ratio = [91, 29, 64, 7]
labels =  ["Majoritarian", "Combined", "PR", "No Directed Elections"]
explode = [0.05, 0.05, 0.05, 0.05]

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True)
plt.show()
plt.savefig('elections_lower_house.png')
```
